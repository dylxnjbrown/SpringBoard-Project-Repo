__Q3:__ Given the nature of the administrator's concern, what is the appropriate confidence interval in this case? A ***one-sided*** or ***two-sided*** interval? (Refresh your understanding of this concept on p. 399 of the *AoS*). Calculate the critical value and the relevant 95% confidence interval for the mean, and comment on whether the administrator should be concerned.

__A: Given the nature of the concern (charges falling below a threshold), a one-sided confidence interval would be appropriate to determine the lower bound of the mean.__

```python
# Calculate statistics
n = len(medical.charges)
m = np.mean(medical.charges)
s = np.std(medical.charges, ddof=1)

# Critical value for 95% confidence (two-tailed)
t_critical = t.ppf(0.975, df=n-1)

# Margin of error
margin_of_error = t_critical * (s / np.sqrt(n))

# Confidence Interval
confidence_interval = (m - margin_of_error, m + margin_of_error)

print(f"95% Confidence Interval: {confidence_interval}")
```

The administrator then wants to know whether people with insurance really are charged a different amount to those without.

__Q4:__ State the null and alternative hypothesis here. Use the _t_-test for the difference between means, where the pooled standard deviation of the two groups is given by:
\begin{equation}
s_p = \sqrt{\frac{(n_0 - 1)s^2_0 + (n_1 - 1)s^2_1}{n_0 + n_1 - 2}}
\end{equation}

and the *t*-test statistic is then given by:

\begin{equation}
t = \frac{\bar{x}_0 - \bar{x}_1}{s_p \sqrt{1/n_0 + 1/n_1}}.
\end{equation}

(If you need some reminding of the general definition of ***t-statistic***, check out the definition on p. 404 of *AoS*). 

What assumption about the variances of the two groups are we making here?

__A: Null Hypothesis (H_0): There is no difference in the average medical charges between patients with insurance and patients without insurance. (\mu{ins} = \mu{no\_ins}) Alternative Hypothesis (H_a): There is a significant difference in the average medical charges between patients with insurance and patients without insurance. (\mu{ins} \neq \mu{no\_ins})__

__Q5:__ Perform this hypothesis test both manually, using the above formulae, and then using the appropriate function from [scipy.stats](https://docs.scipy.org/doc/scipy/reference/stats.html#statistical-tests) (hint, you're looking for a function to perform a _t_-test on two independent samples). For the manual approach, calculate the value of the test statistic and then its probability (the p-value). Verify you get the same results from both.

__A:__ 

```python
drom scipy import stats  # Import the stats module from scipy

# 1. Manual Calculation
# Filter the data into two groups
insured = medical.charges[medical.insuranceclaim == 1]
uninsured = medical.charges[medical.insuranceclaim == 0]

# Calculate n, mean, and variance for each group
n0, n1 = len(insured), len(uninsured)
m0, m1 = np.mean(insured), np.mean(uninsured)
v0, v1 = np.var(insured, ddof=1), np.var(uninsured, ddof=1)

# Calculate pooled standard deviation (sp)
sp = np.sqrt(((n0 - 1) * v0 + (n1 - 1) * v1) / (n0 + n1 - 2))

# Calculate the t-statistic
t_stat_manual = (m0 - m1) / (sp * np.sqrt(1/n0 + 1/n1))

# Calculate the p-value (two-tailed)
p_val_manual = 2 * stats.t.sf(np.abs(t_stat_manual), df=n0 + n1 - 2)

# 2. Scipy Verification
t_stat_scipy, p_val_scipy = stats.ttest_ind(insured, uninsured)

print(f"Manual T-stat: {t_stat_manual}, P-value: {p_val_manual}")
print(f"SciPy T-stat: {t_stat_scipy}, P-value: {p_val_scipy}")
```

Congratulations! Hopefully you got the exact same numerical results. This shows that you correctly calculated the numbers by hand. Secondly, you used the correct function and saw that it's much easier to use. All you need to do is pass your data to it.
