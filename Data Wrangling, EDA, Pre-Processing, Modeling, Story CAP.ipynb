```python
# ============================================================
# INFERENTIAL STATISTICS: Actual Results vs. Expected Skill
# ============================================================

# Define "Luck" as the difference between actual wOBA and expected xwOBA
df2['woba_diff'] = df2['woba'] - df2['xwoba']

# Hypothesis Testing (T-test)
# Null Hypothesis (H0): There is no significant difference between actual wOBA and xwOBA.
# Alternative Hypothesis (Ha): There is a significant gap (indicating league-wide luck/misfortune).

t_stat, p_val = stats.ttest_rel(df2['woba'], df2['xwoba'])

print(f"Paired T-test Results for wOBA vs xwOBA:")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.4f}")

if p_val < 0.05:
    print("Result: Reject the Null Hypothesis. There is a statistically significant difference.")
else:
    print("Result: Fail to reject the Null Hypothesis. Actual results align closely with expected skill.")
```

```python
# ============================================================
#STEP 4: PREPROCESSING AND TRAINING DATA DEVELOPMENT
# ============================================================
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 1. Define all possible features based on your Proposal and Notebook
# We use the column names confirmed in your EDA notebook
target = 'xwoba'
potential_features = [
    'barrel_batted_rate', 'hard_hit_percent', 'avg_best_speed',  # Batted Ball Quality
    'whiff_percent', 'k_percent', 'bb_percent',                 # Plate Discipline
    'xslg', 'xobp',                                             # Expected Baselines
    'ff_avg_speed', 'sl_avg_speed', 'ch_avg_speed'              # Pitch Metrics
]

# FIX: Only use features that are actually in your dataframe (e.g., df2 or df1)
# Replace 'df2' with 'df1' if you haven't created a df2 yet.
active_df = df2 

features = [f for f in potential_features if f in active_df.columns]
print(f"Features found in dataset: {features}")

# 2. Ensure we only use rows where the target and features are not null
df_model = active_df.dropna(subset=[target] + features)

X = df_model[features]
y = df_model[target]

# 3. Standardize numeric features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=features)

# 4. Split into training and testing sets (80/20 split)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled_df, y, test_size=0.20, random_state=42
)

# 5. Review results
print(f"Final training set shape: {X_train.shape}")
```

```python
# ============================================================
# STEP 5: MODELING
# ============================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV

# 1. Initialize models
# We will compare a baseline Linear Regression, a Random Forest, and Gradient Boosting.
lr_model = LinearRegression()
rf_model = RandomForestRegressor(random_state=42)
gb_model = GradientBoostingRegressor(random_state=42)

models = {
    "Linear Regression": lr_model,
    "Random Forest": rf_model,
    "Gradient Boosting": gb_model
}

# 2. Fit and Evaluate Models
model_results = []

for name, model in models.items():
    # Fit the model
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    
    # Calculate metrics
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    
    model_results.append({
        "Model": name,
        "MAE": mae,
        "RMSE": rmse,
        "R2 Score": r2
    })

# Display Results
results_df = pd.DataFrame(model_results).sort_values(by="R2 Score", ascending=False)
print("Initial Model Comparison:")
print(results_df)

# 3. Hyperparameter Tuning (using GridSearchCV on Random Forest)
# We choose Random Forest for tuning as it often provides a strong balance of performance and stability.
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'max_features': ['sqrt', 'log2']
}

print("\nStarting Hyperparameter Tuning for Random Forest...")
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, 
                           cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(X_train, y_train)

best_rf = grid_search.best_estimator_
print(f"Best Parameters: {grid_search.best_params_}")

# 4. Final Evaluation of Best Model
y_pred_final = best_rf.predict(X_test)
final_mae = mean_absolute_error(y_test, y_pred_final)
final_r2 = r2_score(y_test, y_pred_final)

print(f"\nFinal Best Model (Tuned Random Forest) Results:")
print(f"MAE: {final_mae:.4f}")
print(f"R2 Score: {final_r2:.4f}")

# 5. Visualizing Model Performance
plt.figure(figsize=(10, 5))

# Plot 1: Predicted vs Actual
plt.subplot(1, 2, 1)
plt.scatter(y_test, y_pred_final, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Actual xwOBA')
plt.ylabel('Predicted xwOBA')
plt.title('Predicted vs. Actual')

# Plot 2: Feature Importance
plt.subplot(1, 2, 2)
importances = best_rf.feature_importances_
feat_importances = pd.Series(importances, index=X_train.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.title('Top 10 Feature Importances')

plt.tight_layout()
plt.show()
```
